```python
import pandas as pd
import joblib

# Load model & pipeline
model = joblib.load("../models/batsman_runs_model.pkl")
pipeline = joblib.load("../models/featurepipeline.pkl")

# Load new match features
new_data = pd.read_csv("../data/new_match_features.csv")
new_data.head()
```

```python
new_data_processed = pipeline.transform(new_data)
predicted_runs = model.predict(new_data_processed)

new_data["predicted_runs"] = predicted_runs
new_data
```

```python
new_data.to_csv("../data/predicted_runs_output.csv", index=False)
```
